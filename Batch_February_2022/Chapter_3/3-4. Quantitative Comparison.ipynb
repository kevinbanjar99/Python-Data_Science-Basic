<h2>Assignment 3: Preparing Data</h2>

Manipulate that data by grouping it based on its genre, then obtain the mean of each region sales (except the Global_Sales).

```python
df_b = df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
df_b
```

<h2>Assignment 4: Grouping Visualization with Barplot</h2>

1. Visualize the image that we previously manipulated to produce the expected result !
2. Write some insights that you got from that visualization!

```python
import numpy as np

df_c = df_b.transpose()

x = np.arange(0,20,5)
y = []
y_legend = []

for column in df_c.columns :
    y.append(df_c[column])
    y_legend.append(column)

fig, ax = plt.subplots(figsize=(14,9))

for i in range(0,12):
  ax.bar(x-1.3+(i*0.3), y[i], width=0.3, label = y_legend[i])

ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')
ax.set_title('Mean Sales Video Games by Genre')

plt.xticks(ticks=x, labels=list(df_c.index))
plt.style.use('ggplot')
plt.legend()
plt.show()
```

1. Strategy genre has a highest sales in North America
2. Role-Playing genre has a highest sales in Japan
3. Shooter genre has a highest sales in Europe
