````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here

sns.set(style='ticks')


diamonds = sns.load_dataset("diamonds")

sns.despine(fig, left=True, bottom=True)
clarity_order = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
scatter = sns.scatterplot(x="carat", y="price", hue="clarity", size="depth", palette=palette,
          hue_order=clarity_order, sizes=(1, 8), linewidth=0, data=diamonds, ax=ax)
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style("dark"):
  grid = sns.FacetGrid(diamonds, hue='cut', col='cut', height=6)

facet = grid.map(hexbin, "carat", 'price', extent=[0, 3, 0, 20000])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df_rand = df[['cut','carat','depth','table','price']].sample(n=300, random_state=123)

pair = sns.pairplot(df_rand, hue='cut', corner=True, markers=["+", "o", "s", "D", "X"])

#code here
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
from scipy import stats
from warnings import filterwarnings

filterwarnings('ignore')

with sns.axes_style('white') :
  out1 = sns.jointplot("carat", "price", data=df_rand, kind="kde", color="c")

out2 = sns.jointplot("carat", "price", data=df_rand, kind="reg", color="c",
                        xlim=(0, 2.5), ylim=(0, 22000), ).annotate(stats.pearsonr)
````
