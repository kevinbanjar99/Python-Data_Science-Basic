Jawab/Kerjakan pertanyaan/perintah di bawah ini dengan bahasa kalian sendiri:

- Apa itu machine learning?
<p>Machine Learning merupakan cabang ilmu dari ilmu Kecerdasan Buatan, yang fokus pada bagaimana komputer dapat meneningkatkan kecerdasannya dengan belajar dari data-data yang tersedia untuk dapat memprediksi data dimasa depan
- Apa itu data feature dan data target?
<p>Feature = input, berperan sebagai predictor (variabel independent)
<p>Target = output, hasil prediksi (variabel dependent)
- Apa Perbedaan Supervised Learning dan Unsupervised Learning?
<p>Supervised :
<p>a. Menggunakan data latih berupa data yang telah berlabel
<p>b. Bertujuan untuk mengidentifikasi label input yang baru dengan menggunakan fitur prediksi maupun klasifikasi
<p>Unsupervised :
<p>a. Tidak menggunakan data latih, sehingga sistem akan mempelajari data-data yang tidak berlabel
<p>b. Tidak memiliki varibel target, sehingga bertujuan untuk mengelompokkan data-data yang serupa (memiliki kesamaan)
- Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!
<p>a. Classification
Output berupa class label
<p>b. Regression :
Output berupa kuantitas kontinyu
- Apa perbedaan Hyperparameter dan Parameter?
<p>Hyperparameter :
Didefenisikan diawal saat pembuatan objek model
<p>Parameter :
Dihasilkan diakhir
- Sebutkan Langkah-langkah dalam mengaplikasikan algoritma apapun dalam machine learning!
<p> 1) Pemilihan model atau algoritma
<p> 2) Menentukan hyperparameter pada saat membuat objek model
<p> 3) Pemisahan antara data feature dan data target
<p> 4) Eksekusi model untuk mempelajari data menggunakan metode .fit()
<p> 5) Mengaplikasikan model ke data uji coba menggunakan metode .predict() untuk supervised learning atau .transform() dan .predict() untuk unsupervised learning

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# Code here

from sklearn import tree
from  sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_iris

x_reshape = x.reshape(-1,1)
x_new_reshape = x_new.reshape(-1,1)

model_DTR = DecisionTreeRegressor(max_depth=5, min_samples_leaf=0.1, random_state=3)
model_DTR.fit(x_reshape, y)

y_train = model_DTR.predict(x_reshape)
y_test = model_DTR.predict(x_new_reshape)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, color='b', label='Data Asli')
ax.scatter(x,y_train, color='y', label='Prediksi terhadap Data Training')
ax.plot(x_new, y_test, color='g', label='Prediksi terhadap Data Baru')
ax.legend(bbox_to_anchor=(1.1,1.2))
plt.show()
tree.plot_tree(model_DTR);
```
