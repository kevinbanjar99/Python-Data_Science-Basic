- Declare two variables (a & b), where these variables type are numeric
- Declare a variable as the division result between variable a & b, with int datatype
- Declare a variable as the division result between variable a & b, with float datatype


```python
a = 5
b = 3
print("a = " + str(a))
print("b = " + str(b))
print("Data type of variable a = " + str(type(a)))
print("Data type of variable b = " + str(type(a)))

print('')
# int division
c = a/b
c = int(c)
print("c = " + str(c))
print("Data type of variable c " + str(type(c)))

print('')
# float division
d = a/b
d = float(d)
print("d = " + str(d))
print("Data type of variable d " + str(type(d)))
```

<h2>Assignment 4 : String Operation </h2>

- declare your first name into a variable called as 'firstname'
- declare your last name into a variable called as 'lastname'
- Print a text that displayed as : "Hello, my name is *firstname* *lastname* ! I'm ready to learn Python - Basic Data Science"


```python
firstname = "Kevin"
lastname = "Banjarnahor"

# Method 1 : Basic Printing
print("Hello, my name is " + firstname + " " + lastname + " ! I'm ready to learn Python - Basic Data Science")

# Method 2 : Variables (Old Style)
print("Hello, my name is %s %s ! I'm ready to learn Python - Basic Data Science" % (firstname,lastname))

# Method 3 : Variables (New Style)
print("Hello, my name is {0} {1} ! I'm ready to learn Python - Basic Data Science" .format(firstname,lastname))

# Method 4 : String Interpolation
print(f"Hello, my name is {firstname} {lastname} ! I'm ready to learn Python - Basic Data Science")
```

Expected Output :

Hello, my name is Kevin Banjarnahor ! I'm ready to learn Python - Basic Data Science
