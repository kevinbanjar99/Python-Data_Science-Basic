Dataframe is described as two-dimensional object, that basically similiar to 'dictionary' or 'list' that consists of several 'series'. Dataframe has 2 index, and that is row index and column index. Each column of the dataframe consists of 1 series, so each column will only store one data type.

<h2>Assignment 5: Descriptive Statistics</h2>

What is  descriptive statistics ?

Descriptive Statistics is a method that is consists of gathering, clustering, and presentating of a data in order to produce an structured and usable information. Descriptive Statistics Data can be presentated in a form of a graph ( Histogram, Pie Chart, Polygon, Ogive, Stem, and Leaf Diagram ) and numeric ( Central Tendency, Fractile, Skewness, Dispersion)

<h2>Assignment 6: Practice Descriptive Statistics</h2>

- Use a method from pandas to obtain an information about descriptive statistics from a data
- Show the mean value of 'price' column
- Show the standard deviation value of 'depth' column
- Show the maximum value of 'carat' column


```python
# Statistics Descriptive from Pandas

print('INFO : ')
print(df1.info()) # Summary of the dataframe
print('\n')
print('DESCRIPTION : ')
print(df1.describe()) # Stastics Descriptive Conclusion
```

```python
# Mean of price column

print(df1['price'].mean())
```

```python
# Standard deviation of depth column

print(df1['depth'].std())
```

```python
# Maximum of carat column

print(df1['carat'].max())
```
