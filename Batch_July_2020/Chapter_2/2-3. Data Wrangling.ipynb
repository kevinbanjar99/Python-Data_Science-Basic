<h2>Assignment 2: Data Restructuring</h2>

Explain what is the difference between pivoting and melting data?

*   Pivoting is a data reshaping method that penformed by rotating the data that previously placed in rows into columns
*   Melting (also called as unpivoting) is a data reshaping method that penformed by rotating the data that previously placed in columns into rows. To put it simple, melting will make the data that previous wide-shaped become long=shaped.

<h2>Assignment 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='Tipe Ujian', columns='Mata Pelajaran', values=['Rata2 Nilai Ujian','Rata2 Waktu Belajar (menit)'])
pivot_df
```

<h2>Assignment 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data).set_index('Hari')
df
```

```python
df.reset_index(level=0, inplace=True)
df
```

Transform the above data to obtain the expected result :

```python
#code here
#Resetting Index
df.reset_index(level=0, inplace=True)

#Melting
melt_df = pd.melt(pd.DataFrame(data), id_vars = 'Hari', value_vars = ['Kebun Binatang','Dufan','Disney','Bali'], 
                  var_name = 'Tempat Hiburan', value_name = 'Hasil') 
melt_df
```

<h2>Assignment 5: Data Enrichment</h2>

Penform a binning data using qcut to obtain the expected result !

```python
data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}

df = pd.DataFrame(data)
df
```

```python

df['keterangan kelulusan'] = pd.qcut(df['Rata2 Nilai Ujian'], q=3, labels=['Tidak Lulus','Remedial','Lulus'])
df
```
