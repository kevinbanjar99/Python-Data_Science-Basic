**Assignment 4. Time grouping**

 

*   Penform data grouping for the climate changes in 2016 with 1 month frequency and using max anggregation for meantemp, humidity, wind_speed, and meanpressure column !
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
df['2016'].groupby(pd.Grouper(freq='1M')).max()
```

```python
# Highest meantemp, humidity, wind_speed, meanpressure
# Look at rows index 'max'

df['2016'].groupby(pd.Grouper(freq='1M')).max().describe()
```
