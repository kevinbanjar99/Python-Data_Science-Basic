<h2>Assignment 2: Membuat Component Figure dan Axis</h2>

- Create a empty figure and axis !
- Plot codeline below with 2 ways, explicit and implicit!

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

Ans

```python
# Empty figure and axes

import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot()

plt.show()
```

```python
# implicit

plt.plot(x, y)
plt.show()
```

```python
# explicit

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x,y)
plt.show()
```

<h2>Assignment 3: Entry multiple data into one axis</h2>

- Plot these three data into one axis, use seaborn style
- Save the data into a picture file, and then open and view that file.


```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x,y)
ax.plot(x,y2)
ax.plot(x,y3)
plt.style.use('seaborn')
plt.show()
```

```python
fig.savefig('testFig.png')
```
