By using the data below, plot this data and then costumized it to obtain the expected result.

note : used color are green, yellow, blue, and red


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#plt.figure(figsize=(15,20))
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(x,cos,linestyle='solid', color='red', label='Cos x')
ax.plot(x,cos3,linestyle='dashed', color='green', label='3 cos x')
ax.plot(x,sin,linestyle='dashdot', color='blue', label='Sin(x-90)')
ax.plot(x,sin2,marker='x',linestyle='None', color='yellow', label='2 Sin x')

ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')

plt.legend(loc='center', fancybox=True, framealpha=0)
plt.show()
```
