# Assignment 1: Scatter Plot

Create a scatterplot using seaborn from above dataframe, variate it using the provided pallete to produce the expected output.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
 
#code here
fig, ax = plt.subplots()

sns.set(style='ticks')
diamonds = sns.load_dataset("diamonds")

sns.despine(fig, left=True, bottom=True)
clarity_order = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
scatter = sns.scatterplot(x="carat", y="price", hue="clarity", size="depth", palette=palette,
          hue_order=clarity_order.reverse(), sizes=(1, 8), linewidth=0, data=diamonds, ax=ax)
```
