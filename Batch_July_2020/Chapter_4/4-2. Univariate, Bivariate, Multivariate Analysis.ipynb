# Assignment 2. Data Preparation

*   Remove the missing value in the dataset rows using the .dropna() attribute.

*   Penform the data grouping using the 'species' column using the mean() aggregation function. 


```python
#code here

df_edit = df.dropna(axis=0)
df_edit = df_edit.groupby(['species'])[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']].mean()
df_edit
```

# Assignment 3. Univariate Analysis
Create a univariate analysis using these below instructions :

*   Create a Pie Plot from the grouped dataset for flipper_length_mm column to obtain the expected output.
*   Create a histogram plot from the initial dataset for body_mass_g column.
*   Explain the insight that you got from those two graphs. 




```python
#code here
plot = df_edit.plot.pie(y='flipper_length_mm', figsize=(8, 8), legend = False, autopct='%1.1f%%')

plot.axes.xaxis.set_visible(False)
plot.axes.yaxis.set_visible(False)

plt.grid(True)
plt.show()
```
