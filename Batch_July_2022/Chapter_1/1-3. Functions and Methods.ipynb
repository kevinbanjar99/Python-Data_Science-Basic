<h2>Assignment 5: Creating a Simple Function</h2>

Make a function that can process one input in form of a list and return the mean of that list. Give the name of that function as "mean list"


```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50, 13.45, 23.0, 7.0, 8.45, 2.50]

def mean_list(inp_list):
  result = sum(inp_list)/len(inp_list)
  return result

print(mean_list(obj_list))
```

Expected Output:

12.39

# Assignment 7. Mapping Lambda Functiom

Count the amount of even numbers from a list using a lambda function :


```python
obj_list = [i for i in range(1000)]

result = len(list(filter(lambda n : n%2 == 0, obj_list)))

print(result)
```

Expected output:

500
