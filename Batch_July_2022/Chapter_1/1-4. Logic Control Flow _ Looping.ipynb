<h2>Assignment 5: While Loop and For Loop</h2>

- What is the difference between while loop and for loop?
- Show a simple example about the using of while loop dan for loop

While loop is a loop that will always execute its codeline as long as the logical statement value is equal to "true"

syntax ;

while "logical statement" :
<br>codeline

For loop is a loop that will always execute its codeline based on the declared sequence

syntax ;

for "var" in "sequence":
<br>codeline

```python
print('Using while loop: ')
i=0
while i<4 :
  print(i)
  i = i + 1

print('\n')

print('Using for loop: ')
for x in range(0,4):
  print(x)
```

```python
a = -10

def whileFunc(num):
  while num != 0:
    print(num)
    if num > 0 :
      num = num - 1
    else :
      num = num + 1

whileFunc(a)
```

<h2>Soal 7: For Loop</h2>

Obtain the student name that has a score more than or equal to 80. Save that names into 1 list, then show it.







```python
students_data = [
    {
        "name":"Budi",
        "score": 90
    },
    {
        "name":"Nina",
        "score": 78
    },
    {
        "name":"Rudi",
        "score": 91
    },
    {
        "name":"Olivia",
        "score": 76
    },
    {
        "name":"Leo",
        "score": 80
    },
    {
        "name":"Liam",
        "score": 67
    },
    {
        "name":"Sheila",
        "score": 76
    }
]
```

```python
namelist = []

for i in students_data:
  if i['score'] >= 80:
    namelist.append(i['name'])

print(namelist)
```
