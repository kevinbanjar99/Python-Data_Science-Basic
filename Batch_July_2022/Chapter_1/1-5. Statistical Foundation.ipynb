# Assignment 6.

Based from Law of Large Numbers (LNN), a large amount of experiment should be resulted close to the expected value, and will be closer the more experiment that we penform.


![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/e8dff3831b1e5b6f0e187c733583ab8d02574363)

For example, six-sided dice will produce the same probability for each die. So, the estimated obtained mean is equal to ( 1 + 2 + 3 + 4 + 5 + 6 ) / 6 = 3.5. According to LLN, the more dice-throwing that we penform, the result will also closer to the estimated mean, and that is 3.5 .

![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/c/c9/Lawoflargenumbers.svg/400px-Lawoflargenumbers.svg.png)

The same theorem also apply if we have a large dataset of numbers with mean(μ) = 0 and standard deviation (σ) = 1. The more amount of numbers that is in a dataset, the more likely that numbers will become normally distributed.


![alt text](https://img.favpng.com/9/9/7/standard-deviation-normal-distribution-mean-statistics-png-favpng-5cX7qC8aqmp0pQthmW3nSVj9a.jpg)

As displayed from the graph above, if the contents of the dataset become larger, the amount of data that has a value between -1 and 1 equal to 68.2 % from the total data.


Create a codeline to prove that theory !

````python
import numpy as np
from numpy import random # For Dice
from numpy.random import randn

N = 100000                # Sample Size
counter = 0               # Reset the counting process

# Dice

throwResult = random.choice([1,2,3,4,5,6],size=N) # Random Throwing
throwList = list(throwResult) 

print("Mean of the Dice Throwing Result : ")
print(sum(throwList)/len(throwList)) # Mean of the Throwing Result


print("\n")

# Numbers

randomNumber = randn(N) # Generate random distibuted numbers
listNumber = list(randomNumber)

for i in range(0,N) :  
  if listNumber[i] > -1 and listNumber[i] < 1 :          # Start the iteration and add the counter if the condition requirement is met

result = (counter/N)*100         # Calculating the percentage
print("Distribution Percentage of the Numbers: ")
print("%f " % (result))
````
