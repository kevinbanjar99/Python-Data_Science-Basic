<h2>Soal 2: Filter DataFrame with Boolean</h2>

1. Filter the students data with the parental level of educatiom equal to 'high school' or 'bachelor degree'
2. Filter the students data with the 'nilai' value is more than 50 and the test preparation course is completed.
3. Filter the students data with the 'nilai' value is more than 70, the gender is female, and the test preparation course is completed.

```python
1.

HS = df['parental level of education'] == 'high school'
BD = df['parental level of education'] == 'bachelor degree'

df[(HS)|(BD)]
```

```python
2.

score50 = df['nilai'] > 50
testComp = df['test preparation course'] == 'completed'

df[(score50)&(testComp)]
```

```python
3.

score70 = df['nilai'] > 70
genFemale = df['gender'] == 'female'
testComp = df['test preparation course'] == 'completed'

df[(score70)&(testComp)&(genFemale)]
```

<h2>Assignment 3: Data Transformation</h2>

Penfrom an data transformation to the 'nilai' column, with the following transformation function :


![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
NormResult = []

for i in df['nilai']:
  XNorm = (i - df['nilai'].min()) / (df['nilai'].max() - df['nilai'].min())
  NormResult.append(XNorm)

pd.Series(NormResult)
```
