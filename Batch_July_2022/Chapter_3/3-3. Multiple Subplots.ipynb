<h2>Assignment 1: Multiple Subplots</h2>

Create a visualization from the data below to produce the expected output :

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(17,10), sharex=True, sharey=True)

ax[0].plot(x, tan, color='blue', label='tan X')
ax[1].plot(x, sin, color='red', label='sin X')
ax[2].plot(x, cos, color='green', label='cos X')

ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')
plt.show()
```

<h2>Assignment 2: Nested Axis</h2>

With this data, create a visualization to obtain the expected output :


```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax = plt.subplots(figsize=(15,8), sharex=True, sharey=True)
ax.plot(x, y, color='b')
ax.set_title('Grafik CosX')

axul = ax.inset_axes([-0.12, 0.9, 0.3, 0.3])
axul.plot(x,y,color='b')
axul.plot(x,y5,color='y')
axul.set_title('Grafik CosX^5')

axur = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
axur.plot(x,y,color='b')
axur.plot(x,y4,color='k')
axur.set_title('Grafik CosX^4')

axll = ax.inset_axes([-0.12, -0.12, 0.3, 0.3])
axll.plot(x,y,color='b')
axll.plot(x,y2,color='r')
axll.set_title('Grafik CosX^2')

axlr = ax.inset_axes([0.9, -0.12, 0.3, 0.3])
axlr.plot(x,y,color='b')
axlr.plot(x,y3,color='g')
axlr.set_title('Grafik CosX^3')

plt.show()
```
